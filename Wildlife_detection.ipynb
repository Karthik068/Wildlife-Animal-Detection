```python
import os
import torch
from PIL import Image
from torchvision import transforms
from torch.utils.data import Dataset

class WildlifeDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.image_paths = []
        self.labels = []
        self.animal_names = []  # Initialize the list to store animal names
        self.load_data()

    def load_data(self):
        # Loop through each animal's folder
        for animal_folder in os.listdir(self.root_dir):
            animal_folder_path = os.path.join(self.root_dir, animal_folder)
            if os.path.isdir(animal_folder_path):
                print(f"Processing folder: {animal_folder}")  # Debug: show current folder
                self.animal_names.append(animal_folder)  # Store the animal name
                for filename in os.listdir(animal_folder_path):
                    if filename.endswith('.jpg') or filename.endswith('.png'):  # Adjust based on your image formats
                        img_path = os.path.join(animal_folder_path, filename)
                        label_path = os.path.join(animal_folder_path, filename.replace('.jpg', '.txt').replace('.png', '.txt'))

                        if os.path.exists(label_path):
                            self.image_paths.append(img_path)
                            with open(label_path, 'r') as f:
                                bboxes = []
                                # Read and parse all lines in the label file
                                for line in f:
                                    data = line.strip().split()
                                    # Ensure there are 4 bounding box coordinates
                                    if len(data) == 4:  # 4 coordinates
                                        bbox = list(map(float, data))  # Convert all elements to float
                                        bboxes.append(bbox)  # Collect bounding boxes

                                if bboxes:
                                    # Convert list of bounding boxes to a tensor
                                    self.labels.append({
                                        'boxes': torch.tensor(bboxes, dtype=torch.float32),  # Shape: [N, 4]
                                        'labels': torch.tensor([self.animal_to_index(animal_folder)] * len(bboxes), dtype=torch.int64)  # Labels as tensors
                                    })
                                else:
                                    print(f"No valid bounding boxes in file {label_path}. Skipping this image.")
                                    # If no bounding boxes are found, skip the corresponding image
                                    self.image_paths.pop()  # Remove the last image added (since it has no valid boxes)
                        else:
                            print(f"Label file not found for {img_path}. Skipping this image.")

        print(f"Total images: {len(self.image_paths)}")  # Debug: Show total images loaded
        print(f"Total targets: {len(self.labels)}")  # Debug: Show total targets loaded

    def animal_to_index(self, animal_name):
        """Convert animal folder name to an index for labels."""
        return self.animal_names.index(animal_name)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        target = self.labels[idx]

        image = Image.open(img_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        return image, target  # Return image and target dict


# Define transformations (adjusted according to your needs)
transform = transforms.Compose([
    transforms.Resize((980, 780)),  # Resize images to the input size for Faster R-CNN
    transforms.RandomHorizontalFlip(p=0.5),  # Randomly flip images with a probability of 0.5
    transforms.RandomRotation(degrees=15),  # Randomly rotate images by ±15 degrees
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # Randomly adjust colors
    transforms.ToTensor(),
])

# Create dataset instance
dataset = WildlifeDataset(root_dir='/content/AnimalImages1/Images', transform=transform)  # Update the path accordingly

# Example to get a sample
for i in range(len(dataset)):  # Iterate over all samples in the dataset
    image, target = dataset[i]
    print(f"Image shape: {image.shape}")
    print(f"Target: {target}\n")  # Print targets
```

```python
import tensorflow as tf
import os
from PIL import Image
import io

def create_tfrecord_example(image_path, bbox, class_label):
    # Read the image
    with open(image_path, 'rb') as img_file:
        img_data = img_file.read()

    # Get image dimensions
    image = Image.open(io.BytesIO(img_data))
    width, height = image.size

    # Convert bounding box to relative coordinates
    x_min, y_min, x_max, y_max = bbox
    x_min /= width
    x_max /= width
    y_min /= height
    y_max /= height

    # Create TFRecord example
    feature = {
        'image': tf.train.Feature(bytes_list=tf.train.BytesList(value=[img_data])),
        'bbox': tf.train.Feature(float_list=tf.train.FloatList(value=[x_min, y_min, x_max, y_max])),
        'label': tf.train.Feature(int64_list=tf.train.Int64List(value=[class_label]))
    }
    return tf.train.Example(features=tf.train.Features(feature=feature))

def save_tfrecords(output_path, dataset):
    with tf.io.TFRecordWriter(output_path) as writer:
        for image_path, bbox, class_label in dataset:
            example = create_tfrecord_example(image_path, bbox, class_label)
            writer.write(example.SerializeToString())

# Example usage
dataset_path = '/content/AnimalImages1/Images'  # Main directory with subfolders for each animal
tfrecord_path = 'wildlife_dataset.tfrecord'

# Define a mapping from folder names to integer class labels
# Example mapping, adjust according to your dataset structure
folder_to_class_index = {
    'Lion': 1,
    'Jaguar': 2,
    'Cheetah': 3,
    'Tiger': 4,
    'Leopard': 5
    # Add other animal names with corresponding indices
}

dataset = []
# Loop through each animal folder
for folder_name in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, folder_name)
    if not os.path.isdir(folder_path):
        continue

    # Assign class label based on folder name
    class_label = folder_to_class_index.get(folder_name)
    if class_label is None:
        raise ValueError(f"Folder name '{folder_name}' does not have a corresponding class index.")

    # Loop through each file in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.jpg') or file_name.endswith('.png'):
            image_path = os.path.join(folder_path, file_name)

            # Assuming label file is named similarly to image file but with .txt extension
            label_path = os.path.splitext(image_path)[0] + '.txt'
            if os.path.exists(label_path):
                # Read bounding box from the label file, assuming only bounding box coordinates are present
                with open(label_path, 'r') as f:
                    content = f.readline().split()
                    # Now, only read the bounding box coordinates, assuming they start from content[0]
                    bbox = list(map(float, content[:4]))  # Adjust this based on the format in your label file

                # Append to dataset
                dataset.append((image_path, bbox, class_label))

# Save all samples to TFRecord
save_tfrecords(tfrecord_path, dataset)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import tensorflow as tf
import numpy as np

class TFRecordDataset(Dataset):
    def __init__(self, tfrecord_path):
        self.raw_dataset = tf.data.TFRecordDataset(tfrecord_path)
        self.parsed_dataset = self.raw_dataset.map(parse_tfrecord_fn)

    def __len__(self):
        return sum(1 for _ in self.raw_dataset)  # Length of the dataset

    def __getitem__(self, idx):
        for image, bbox, label in self.parsed_dataset.skip(idx).take(1):  # Select a single item
            # Convert to PyTorch tensors
            image_tensor = torch.from_numpy(image.numpy()).permute(2, 0, 1).float()  # HWC to CHW format
            bbox_tensor = torch.tensor([bbox.numpy()], dtype=torch.float32)  # Bounding box tensor
            label_tensor = torch.tensor([label.numpy()], dtype=torch.int64)  # Class label tensor

            target = {'boxes': bbox_tensor, 'labels': label_tensor}
            return image_tensor, target
```

```python
# Load dataset function
def load_dataset(tfrecord_path, batch_size=32):
    # Define the feature description for the TFRecord
    feature_description = {
        'image': tf.io.FixedLenFeature([], tf.string),
        'bbox': tf.io.FixedLenFeature([4], tf.float32),
        'label': tf.io.FixedLenFeature([], tf.int64),
    }

    def _parse_function(example_proto):
        # Parse the input `tf.train.Example` proto using the above feature description
        parsed_features = tf.io.parse_single_example(example_proto, feature_description)

        # Decode the image data
        image = tf.io.decode_jpeg(parsed_features['image'], channels=3)
        image = tf.image.resize(image, [980, 780])  # Resize to model input shape
        image = tf.cast(image, tf.float32) / 255.0  # Normalize to [0, 1]

        # Get bounding box and label
        bbox = parsed_features['bbox']
        label = parsed_features['label']

        return image, (label, bbox)

    # Create a dataset from the TFRecord file
    dataset = tf.data.TFRecordDataset(tfrecord_path)
    dataset = dataset.map(_parse_function)
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)

    return dataset

# Set file path and batch size
tfrecord_path = '/content/wildlife_dataset.tfrecord'  # Path to your TFRecord file
batch_size = 32

# Load the dataset
train_dataset = load_dataset(tfrecord_path, batch_size=batch_size)

# Check the dataset output
for images, labels in train_dataset.take(1):
    print(f"Image batch shape: {images.shape}")
    print(f"Labels: {labels}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# Set parameters
img_height, img_width = 980, 780
num_classes = 5  # Number of classes in your dataset

# Custom region proposal network (simplified)
def region_proposal_network(feature_map):
    x = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(feature_map)
    objectness = layers.Conv2D(1, (1, 1), activation='sigmoid')(x)  # object/background classification
    bbox_deltas = layers.Conv2D(4, (1, 1))(x)  # bounding box regression
    return objectness, bbox_deltas

# Custom classifier and regressor head
def classifier_and_regressor_head(features, num_classes):
    x = layers.Flatten()(features)
    x = layers.Dense(1024, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    class_logits = layers.Dense(num_classes, activation='softmax')(x)
    bbox_deltas = layers.Dense(4)(x)  # bounding box adjustment
    return class_logits, bbox_deltas

# Define the RCNN model
def rcnn_model(input_shape=(980, 780, 3)):
    inputs = layers.Input(shape=input_shape)

    # Feature extraction layers
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)
    x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)
    x = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = layers.GlobalAveragePooling2D()(x)

    # Classification head
    class_output = layers.Dense(5, activation='softmax', name='class_output')(x)

    # Bounding box regression head
    bbox_output = layers.Dense(4, name='bbox_output')(x)

    model = models.Model(inputs=inputs, outputs=[class_output, bbox_output])

    return model

# Create and compile the model
model = rcnn_model()
model.compile(optimizer='adam',
              loss={'class_output': 'sparse_categorical_crossentropy', 'bbox_output': 'mse'},
              metrics={'class_output': 'accuracy', 'bbox_output': 'mse'})

# Print model summary
model.summary()
```

```python
# Train the model
epochs = 20  # Adjust as needed
history = model.fit(
    train_dataset,
    epochs=epochs
)

# Save the model
model.save('wildlife_rcnn_model.h5')
```
